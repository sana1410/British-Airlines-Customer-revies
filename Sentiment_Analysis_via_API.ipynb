```python
from google.cloud import storage
storage_client=storage.Client()
files_list=storage_client.list_blobs("eai6010_model_02122024")
for files in files_list:
  print(files.name)
bucket=storage_client.bucket("eai6010_model_02122024")
print(bucket)
blob=bucket.blob("Customer reviews.csv/Reviews.csv")
print(blob)
blob.download_to_filename('/content/Customer reviews.csv')
```

```python
import pandas as pd
df= pd.read_csv('/content/Customer reviews.csv')
```

```python
df.head()
```

```python
from google.cloud import language_v2
sentiment_score=[]
sentiment_magnitude=[]
for i in range(len(df)):
    client = language_v2.LanguageServiceClient()
    text=df['REVIEWS'][i]
    print(f'row number ',{i})





    # gcs_content_uri = 'gs://cloud-samples-data/language/entity-sentiment.txt'

    # Available types: PLAIN_TEXT, HTML
    type_ = language_v2.Document.Type.PLAIN_TEXT

    # Optional. If not specified, the language is automatically detected.
    # For list of supported languages:
    # https://cloud.google.com/natural-language/docs/languages
    language_code = "en"
    document = {
        "content": text,
        "type_": type_,
        "language_code": language_code,
    }

    # Available values: NONE, UTF8, UTF16, UTF32
    encoding_type = language_v2.EncodingType.UTF8

    response = client.analyze_sentiment(
        request={"document": document, "encoding_type": encoding_type}
    )
    # Get overall sentiment of the input document
    print(f"Document sentiment score: {response.document_sentiment.score}")
    sentiment_score.append(response.document_sentiment.score)
    print(f"Document sentiment magnitude: {response.document_sentiment.magnitude}")
    sentiment_magnitude.append(response.document_sentiment.magnitude)

```

```python
print(len(sentiment_score))
```

```python
print(sentiment_score)
print(sentiment_magnitude)
```

```python
df.info()
```

```python
sentiment_df=pd.DataFrame(sentiment_score,columns=["sentiment_score"])
```

```python
sentiment_magnitude=pd.DataFrame(sentiment_magnitude,columns=["sentiment_magnitude"])
```

```python
sentiment_df=pd.concat([sentiment_df,sentiment_magnitude],axis=1)
```

```python
sentiment_df.info()
```

```python
df_final=pd.concat([df,sentiment_df],axis=1)
```

```python
df_final.info()
```

```python
# Convert the DataFrame to a CSV string
csv_data = df_final.to_csv(index=False)

# Get the bucket
bucket = storage_client.bucket('eai6010_model_02122024')

# Get the blob (file) in the bucket
blob = bucket.blob('Customer sentiment.csv')

# Upload the CSV data to the GCS file
blob.upload_from_string(csv_data, content_type='text/csv')
```
